## Countries with Number of Billionaries

```python
#
df['Country']
```

```python
#
for country in df['Country'].unique():        # Taking 1 country at a time
    cnt = 0
    for i in df['Country']:                   # Find number of Billionaries in that specific country
        if(i == country):
            cnt += 1
    print([country,cnt])        
```

```python
# Storing it in one list
data = []
for country in df['Country'].unique():        # Taking 1 country at a time
    cnt = 0
    for i in df['Country']:                   # Find number of Billionaries in that specific country
        if(i == country):
            cnt += 1
    data.append([country,cnt]) 
print(data)    
```

```python
# Creating a dataframe from the data

data = pd.DataFrame(data,columns = ['Country','Frequency'])
```

```python
# Top 5 Elements
data.head()
```

```python
# Sorting the data
data.sort_values(by = 'Frequency')
```

```python
# Sorting the data in deascending order
data.sort_values(by = 'Frequency', ascending = False)
```

Here you can see that the datatype of the "NetWorth" column is Object not a "Integer". So we have to change it.

```python
# Store all values in new list after converting it into Integer
net_worth = []

for i in df['NetWorth']:
    net_worth.append(int(float(i[1:-2])))
    
```
